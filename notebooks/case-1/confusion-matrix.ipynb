```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_excel("C:/Users/dbaeva/Documents/Docs/master-probability-causality/data/case-1/SO-observ.xlsx", 
                   sheet_name="comparison-table")  
```

```python
# Mapping labels to numerical values
actual = df['SO-actual'].map({'R': 1, 'G': 0})  
predicted = df['TPMS-predicted'].map({'R': 1, 'G': 0})  

# Generate confusion matrix
matrix = confusion_matrix(actual, predicted, labels=[1, 0])  # [1, 0] for Red, Green

# Plot the confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(matrix, annot=True, fmt='d', cmap='Blues',
           xticklabels=['Predicted Red', 'Predicted Green'],
           yticklabels=['Actual Red', 'Actual Green'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
# Calculate accuracy from confusion matrix
accuracy_from_matrix = (matrix[0, 0] + matrix[1, 1]) / matrix.sum()
print(f"Accuracy from confusion matrix: {accuracy_from_matrix * 100:.2f}%")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Calculate precision
precision = precision_score(actual, predicted)
print(f"Precision: {precision:.2f}")

# Calculate recall
recall = recall_score(actual, predicted)
print(f"Recall: {recall:.2f}")

# Calculate F1 score
f1 = f1_score(actual, predicted)
print(f"F1 Score: {f1:.2f}")
```
